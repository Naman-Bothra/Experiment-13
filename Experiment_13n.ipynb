```python
# Preprocessing on da Dataset-1
import pandas as pd
data={
    'Roll No':[101,102,103,104,105],'Name':["Asha","Rohan","Meera","Karan","Anil"],
    "Age": [20, "-", 21, 20, "22"],
    "Marks": [85, 90, "-", 88, 76],
    "Department": ["CSE", "IT", "CSE", "cse", "E&TC"],
    "Admission_Date": ["01-06-2023", "03-06-2023", "05-06-2023", "07-06-2023", "10/06/23"]
}
df = pd.DataFrame(data)
df
```

```python
# replace '-' with Nan to mark missing values
df.replace("-",np.nan,inplace=True)

# Convert columns to numeric where needed
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Marks']= pd.to_numeric(df['Marks'], errors='coerce')
df
```

```python
# Check Missing values
print("\n Missing Values:")
print(df.isna().sum())
```

```python
# fill missing Age with mean
df['Age'].fillna((df['Age'].mean()), inplace=True)
df['Marks'].fillna(df['Marks'].median(), inplace=True)
df
```

```python
# Fix inconsistent text formatting
df["Department"]=df["Department"].str.upper()
df
```

```python
# Convert admission date to standard date time format:
df['Admission_Date'] = pd.to_datetime(df['Admission_Date'],errors='coerce',dayfirst=True)
print(df)
''' If a Value cannot be converted to a valid date, Pandas will repace it with NaT(Not a Time)
Eg: invalid dates like "abc" or "32-13-2024" becomes NaT'''
```
